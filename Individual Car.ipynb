```python
import time
import pandas as pd
import csv 
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from msedge.selenium_tools import Edge, EdgeOptions
```

```python
options  = EdgeOptions()
options.use_chromium = True
driver = Edge(options=options)
```

```python
lst = []
```

```python
def pass_link(url):
    specs = {}
    all_features = []
    driver.get(url)
    soup = BeautifulSoup(driver.page_source,'html.parser')
    title = soup.find('h1',{'id':'content-title'})
    div = soup.find('div',{'class':'overview-list'})
    try:
        specs['Car Title'] = title.text.strip()
    except:
        return
    target = div.find_all('div')
    for i in range(0,len(target),3):
        specs[target[i].text.strip()] = target[i+1].text.strip()
    driver.find_element(By.XPATH,"//li[text()='Specifications']").click()
    soup = BeautifulSoup(driver.page_source,'html.parser')

    
    div = soup.find('div',{'class':'overview-list'})
    try:
        driver.find_element(By.XPATH,"//li[text()='Specifications']").click()
    except:
        return
    div = soup.find_all('div',{'id':'specifications'})
    for i in div:
        target_div = i.find_all('div',{'class':'description'})
        for k in range(len(target_div)):
            target = target_div[k].find_all('div',{'class':'property-wrapper'})
            for j in range(0,len(target)):
                header = str(target[j].text).split('\n')[0].replace('\t','')
                details = str(target[j].text).split('\n')[-1].replace('\t','')
                specs[header] = details
    try:
        driver.find_element(By.XPATH,"//li[text()='Features']").click()      
        soup = BeautifulSoup(driver.page_source,'html.parser')
        div = soup.find_all('div',{'id':'features'})
        for i in div:
            target_div = i.find_all('div',{'class':'description'})
            for k in range(len(target_div)):
                target = target_div[k].find_all('div',{'class':'property-wrapper'})
                for j in target:
                    div_final = j.find_all('div')
                    for l in div_final:
                        if str(l.text.strip())[0:2]=='No':
                            pass
                        else:
                            all_features.append(l.text.strip())
        specs['Features'] = all_features
    except:
        pass
    
    lst.append(specs)
```

```python
one = pd.read_csv('Aagra_car_links.csv')
two = pd.read_csv('Amritsar_car_links.csv')
three = pd.read_csv('Chandigarh_car_links.csv')
four = pd.read_csv('Jaipur_car_links.csv')
five = pd.read_csv('Kanpur_car_links.csv')
six = pd.read_csv('Koimbatur_car_links.csv')
seven = pd.read_csv('Lucknow_car_links.csv')
eight = pd.read_csv('Ludhiana_car_links.csv')
nine = pd.read_csv('Mohali_car_links.csv')
ten = pd.read_csv('Panchkula_car_links.csv')
eleven = pd.read_csv('Patiala_car_links.csv')
twelve = pd.read_csv('Unnao_car_links.csv')
```

```python
df = pd.concat([one,two,three,four,five,six,seven,eight,nine,ten,eleven,twelve],ignore_index=True)
```

```python
df
```

```python
for i in range(3344):
    pass_link(df['link'][i])
    driver.implicitly_wait(5)
    print('car '+str(i)+' done')
```
